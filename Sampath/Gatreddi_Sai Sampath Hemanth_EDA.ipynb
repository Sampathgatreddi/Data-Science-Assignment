```python
max_transactions_customer = list(customers_data[customers_data['CustomerID'] == max(transactions_data['CustomerID'])]['CustomerName'])
```

```python
max_customer_region = list(customers_data[customers_data['CustomerName'] == max_transactions_customer[0]]['Region'])
```

```python
print(f"The maximum number of transactions are done by customer  {max_transactions_customer[0]} and is from {max_customer_region[0]}.")
```

```python
print(f"The maximum number of customers are from {max(customers_data['Region'])}.")
```

```python
max_transaction_value = transactions_data['TotalValue'].max()
largestCustomer = customers_data[customers_data['CustomerID'].isin(transactions_data[transactions_data['TotalValue'] == max_transaction_value]['CustomerID'])]['CustomerID'].tolist()
largestCustomerName = list(customers_data[customers_data['CustomerID'] == largestCustomer[0]]['CustomerName'])
largestCustomerRegion = list(customers_data[customers_data['CustomerID'] == largestCustomer[0]]['Region'])
```

```python
print(f"The largest Transaction is about {max(transactions_data['TotalValue'])} and is done by {largestCustomerName[0]} and is from {largestCustomerRegion[0]}.")
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
customers_data['Region'] = le.fit_transform(customers_data['Region'])
customers_data['ID'] = le.fit_transform(customers_data['CustomerID'])
```

```python
transactions_data = transactions_data.merge(
    customers_data[['CustomerID', 'ID']],  # Select the relevant columns from customers_data
    on='CustomerID',                      # Match on CustomerID
    how='left'                            # Keep all rows in transactions_data
)
```
